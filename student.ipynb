```python
# Calculate the correlation matrix
corr = df.corr()

# Set a threshold for identifying highly correlated features
threshold = 0.7

# Identify highly correlated features
multicollinear_features = []

def find_multicollinear_features(feature):
    multicollinear = corr[feature][(corr[feature] >= threshold) & (corr.index != feature)]
    if not multicollinear.empty:
        multicollinear_features.extend([(feature, idx, round(corr[feature][idx], 2)) for idx, _ in multicollinear.items()])

# Loop through features to find highly correlated pairs
for feature in corr:
    find_multicollinear_features(feature)

# Create a DataFrame of highly correlated features
MC_df = pd.DataFrame(multicollinear_features, columns=['Feature 1', 'Feature 2', 'Correlation'])
print('Highly Correlated Features')
display(MC_df)
```

We have seen that sqft living and sft above have the higest collinearity among them so we will pick sqft living as our baseline and drop sqft above to prevent multicollinearity in our model.

### Multiple Regression model 1

```python
# Multiple Regression Diagnostic
details = "bedrooms + bathrooms + sqft_living +  sqft_lot + floors + sqft_above"
formula =  "price~" + details
model = sm.OLS.from_formula(formula, data=hse_reg)
results = model.fit()
print(results.summary())
# model.summary()
```

There was a high level of skewness to the right(2.781) from the data. We want to remedy this by normalizing the data. To do this we decided to log transform the variables.

```python
# Regression Diagnostic
newX = df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','sqft_above']]
newX
fig = plt.figure(figsize=(15, 8))
sm.graphics.plot_partregress_grid(results, exog_idx= list(newX.columns.values), grid=(3,4), fig=fig)
plt.show()
```

In the multilinear regression model, the residuals were not evenly spread out. This also made the model unreliable.

```python
#Due to skewness we decided to log transform
y_log = np.log(hse_reg["price"])
sns.distplot(y_log, fit=norm)
plt.show()
```

After using log scaling we normalised the price data makin it more usuable in multilinear regression.
